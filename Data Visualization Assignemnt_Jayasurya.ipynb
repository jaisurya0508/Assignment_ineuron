```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

### Load iris dataset from sklearn

```python
iris_data=datasets.load_iris()
```

```python
iris_data
```

```python
x=PCA(n_components=3).fit_transform(iris_data.data)
y=iris_data.target
```

```python
x[0:5]
```

```python
y
```

```python
fig=plt.figure(1, figsize=(8,6))
ax=Axes3D(fig, elev=-150, azim=110)
ax.scatter(x[:,0],x[:,1],x[:,2],c=y,cmap=plt.cm.rainbow_r)
```

```python
ax.set_title("Three PCA")
ax.set_xlabel("first eigen vector")
ax.set_ylabel("second eigen vector")
ax.set_zlabel("third eigen vector")
```
