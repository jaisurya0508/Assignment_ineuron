### In this assignment students have to transform iris data into 3 dimensionsand plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
x_reduced=PCA(n_components=3).fit_transform(iris.data)
y=iris.target
```

```python
fig=plt.figure(1,figsize=(8,6))
ax=Axes3D(fig, elev=-150,azim=110)
ax.scatter(x_reduced[:,0],x_reduced[:,1],x_reduced[:,2],c=y,cmap=plt.cm.rainbow_r)
ax.set_title("First Three PCA directions")
ax.set_xlabel("1st eigen vector")
ax.set_ylabel("2nd eigen vector")
ax.set_zlabel("3rd eigen vector")
plt.show()
```
